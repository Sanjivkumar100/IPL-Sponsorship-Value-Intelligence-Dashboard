```python
from google.colab import userdata

api_key = userdata.get("YOUTUBE_API_KEY")
```

```python
from googleapiclient.discovery import build

youtube = build('youtube', 'v3', developerKey=api_key)
```

```python
team_data=[]
for team in teams:
  print(f"Fetching data for {team}")
  request=youtube.search().list(
      q=f'{team}ipl highlights',
      part='snippet',
      type='video',
      maxResults=20
  )
  response= request.execute()

  video_ids=[]
  channel_ids=set()
  for item in response['items']:
    video_ids.append(item['id']['videoId'])
    channel_ids.add(item['snippet']['channelId'])

  stats_request=youtube.videos().list(
      part="statistics",
      id=",".join(video_ids)
  )
  stats_response=stats_request.execute()

  for item in stats_response['items']:
    stats=item['statistics']

    views=int(stats.get('viewCount',0))
    likes=int(stats.get('likeCount',0))
    comments=int(stats.get('commentCount',0))

    team_data.append([team,views,likes,comments])

df=pd.DataFrame(team_data,columns=['team','views','likes','comments'])

df['views']=df['views'].fillna(0)
df['likes']=df['likes'].fillna(0)
df['comments']=df['comments'].fillna(0)
df['engagement']=((df['likes']+df['comments'])/df['views'])*100
df['engagement'] = df['engagement'].fillna(0)

teams_engagement = df.groupby('team').agg({
    'views':'mean',
    'likes':'mean',
    'comments':'mean',
    'engagement':'mean'
}).reset_index()

teams_engagement
```

```python
def get_popularity(row):
  t1=row['team1']
  t2=row['team2']
  return(row.get(t1,0)+row.get(t2,0))/2
merged['popularity_score']=merged.apply(get_popularity,axis=1)
```

```python
# Merge team1
final_df = pd.merge(
    merged,
    teams_engagement,
    left_on='team1',
    right_on='team',
    how='left'
)

# Merge team2
final_df = pd.merge(
    merged,
    teams_engagement,
    left_on='team2',
    right_on='team',
    how='left',
    suffixes=('_team1', '_team2')
)
```

```python
final_df = final_df[
    final_df['team1'].isin(teams_engagement['team']) &
    final_df['team2'].isin(teams_engagement['team'])
]
```

```python
final_df = final_df.drop_duplicates()
```

```python
final_df.columns
```

```python
# Merge team1
final_df = pd.merge(
    final_df,
    teams_engagement,
    left_on='team1',
    right_on='team',
    how='left'
)

# Merge team2
final_df = pd.merge(
    final_df,
    teams_engagement,
    left_on='team2',
    right_on='team',
    how='left',
    suffixes=('_team1', '_team2')
)
```

```python
final_df['match_engagement']=((final_df['engagement_x']+final_df['engagement_y'])/2)
```

```python
final_df[['team1', 'team2', 'match_engagement']]
```

```python
final_df.rename(columns={
    'views_x': 'views_team1',
    'views_y': 'views_team2',
    'engagement_x': 'engagement_team1',
    'engagement_y': 'engagement_team2'
}, inplace=True)
```

```python
final_df['SVI'] = (
    0.4 * final_df['attendance_proxy'] +
    0.3 * final_df['popularity_score'] +
    0.3 * final_df['match_engagement']
)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

final_df[['attendance_proxy',
          'popularity_score',
          'match_engagement']] = scaler.fit_transform(

    final_df[['attendance_proxy',
              'popularity_score',
              'match_engagement']]
)
```
